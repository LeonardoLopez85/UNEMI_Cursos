```python
def generador_pares(limite):
    num = 0
    while num < limite:
        yield num
        num += 2

for par in generador_pares(10):
    print(par)
```

```python
def contar_hasta(n):
    for i in range(n):
        yield i

for numero in contar_hasta(5):
    print(numero)
```

```python
gen = (x**2 for x in range(10))

for valor in gen:
    print(valor)
```
